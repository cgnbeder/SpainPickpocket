```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
# Imported data immigrations data
eurostat_immigrations = pd.read_excel("datas/eurostat_immigrations.xlsx")     
eurostat_immigrations.head(10)
```

```python
# Cleaned data and defined it as a clean one. From here, I'm going to use the clean_eurostat_i dataframe
clean_eurostat_i = eurostat_immigrations.drop(range(0, 4))
clean_eurostat_i.replace(':', pd.NA, inplace=True)
clean_eurostat_i = clean_eurostat_i.dropna(subset=['TIME'])
clean_eurostat_i.reset_index(drop=True, inplace=True)
clean_eurostat_i.rename(columns={"TIME": "Country Name"}, inplace=True)
clean_eurostat_i
```

```python
# Uploaded World Bank Data unemployment dataset
unemployment_wb = pd.read_excel("datas/unemployment_wb.xlsx")

# Column names have the same year twice. So I dropped all characters after the first 4 digits to clean column names.
# The first column and its values should not be affected.

unemployment_wb.rename(columns=lambda x: x[:4] if len(x) > 4 and x != "Country Name" else x, inplace=True)
unemployment_wb
```

```python
# SECOND unemployment dataset (Eurostats)
unemployment_es = pd.read_excel("datas/unemployment_es.xlsx")
unemployment_es.rename(columns={"GEO (Labels)": "Country Name"}, inplace=True)
unemployment_es = unemployment_es.drop(range(0, 1))
unemployment_es
```

```python
# Imported Spain tourist number per year
spain_tourist_number = pd.read_excel("datas/Spain_tourist_number.xlsx")
spain_tourist_number
```

```python
# Imported Spain tourists' regions per year (if necessary)
spain_tourist_number_by_region = pd.read_excel("datas/Spain_tourist_region.xlsx")
spain_tourist_number_by_region
```

```python
# Imported police-recorded offences by country (number)
police_recorded_offences_number = pd.read_excel("datas/Police_recorded_offences_number.xlsx")
police_recorded_offences_number.rename(columns={"GEO (Labels)": "Country Name"}, inplace=True)
police_recorded_offences_number
```

```python
# Imported police-recorded offences by country (Per hundred thousand inhabitants)
police_recorded_offences_pht = pd.read_excel("datas/Police_recorded_offences_pht.xlsx")
police_recorded_offences_pht.rename(columns={"GEO (Labels)": "Country Name"}, inplace=True)
police_recorded_offences_pht
```

# Dataset Summary
- clean_eurostat_i = immigration numbers by country https://ec.europa.eu/eurostat
- unemployment_es = unemployment rate by country (Eurostat data) https://ec.europa.eu/eurostat/databrowser/view/tps00203/default/table?lang=en
- unemployment_wb = unemployment rate by country (World Bank dataset) https://data.worldbank.org/indicator/SL.UEM.TOTL.ZS
- spain_tourist_number = number of tourists by country. Filtered only Spain before importing https://www.unwto.org/tourism-statistics/key-tourism-statistics
- spain_tourist_number_by_region = Spain tourist numbers by region https://www.unwto.org/tourism-statistics/key-tourism-statistics
- police_recorded_offences_number = police-recorded offences number by country https://ec.europa.eu/eurostat/databrowser/view/crim_off_cat/default/table?lang=en
- police_recorded_offences_pht = police-recorded offences by country per hundred thousand inhabitants https://ec.europa.eu/eurostat/databrowser/view/crim_off_cat/default/table?lang=en

```python
# Now, I'm going to merge all tables and filter only Spain, then melt them to use years as columns and rows as categories.
spain_eurostat_immigrations = clean_eurostat_i[clean_eurostat_i["Country Name"] == "Spain"]
spain_eurostat_immigrations

spain_unemployment_wb = unemployment_wb[unemployment_wb["Country Name"] == "Spain"]
spain_unemployment_wb.drop(columns=["Coun"], inplace=True)
spain_unemployment_wb

spain_unemployment_es = unemployment_es[unemployment_es["Country Name"] == "Spain"]
spain_unemployment_es

spain_tourist_number_total = spain_tourist_number[spain_tourist_number["Arrivals"] == "Total arrivals"]
spain_tourist_number_total

spain_police_recorded_offences_number = police_recorded_offences_number[police_recorded_offences_number["Country Name"] == "Spain"]
spain_police_recorded_offences_number
```

Because all tables have different year gaps, I filtered from 2014 to 2022. This way, there will be no null values or empty spaces in our merged, melted table.

```python
spain_eurostat_immigrations  = spain_eurostat_immigrations.iloc[:, 4:]
spain_eurostat_immigrations["Country Name"] = "Spain"
spain_eurostat_immigrations
```

```python
spain_unemployment_wb = spain_unemployment_wb.iloc[:, 3:]
spain_unemployment_wb.drop(columns=["2023"], inplace=True)
spain_unemployment_wb["Country Name"] = "Spain"
spain_unemployment_wb
```

```python
spain_unemployment_es  = spain_unemployment_es.iloc[:, 3:]
spain_unemployment_es.drop(columns=["2023"], inplace=True)
spain_unemployment_es["Country Name"] = "Spain"
spain_unemployment_es
```

```python
spain_tourist_number_total = spain_tourist_number_total.iloc[:, 21:]
spain_tourist_number_total["Country Name"] = "Spain"
spain_tourist_number_total = spain_tourist_number_total.applymap(lambda x: x.strip() if isinstance(x, str) else x)
spain_tourist_number_total
```

```python
spain_police_recorded_offences_number = spain_police_recorded_offences_number.iloc[:, 7:]
spain_police_recorded_offences_number["Country Name"] = "Spain"
spain_police_recorded_offences_number
```

```python
# I filtered Spain from all tables. Now, I'm going to merge all of them and melt them to use years as columns and categories as rows.
spain_eurostat_immigrations_melted = spain_eurostat_immigrations.melt(id_vars=["Country Name"], 
                                                                      var_name="Year", 
                                                                      value_name="Immigrations")
spain_unemployment_wb_melted = spain_unemployment_wb.melt(id_vars=["Country Name"], 
                                                          var_name="Year", 
                                                          value_name="Unemployment WB")
spain_unemployment_es_melted = spain_unemployment_es.melt(id_vars=["Country Name"], 
                                                          var_name="Year", 
                                                          value_name="Unemployment ES")
spain_tourist_number_melted = spain_tourist_number_total.melt(id_vars=["Country Name"], 
                                                              var_name="Year", 
                                                              value_name="Tourist Arrivals")
spain_offences_number_melted = spain_police_recorded_offences_number.melt(id_vars=["Country Name"], 
                                                                          var_name="Year", 
                                                                          value_name="Recorded Offences")

# Dropping the "Country Name" column, we will merge using the "Year" column
spain_eurostat_immigrations_melted.drop(columns=["Country Name"], inplace=True)
spain_unemployment_wb_melted.drop(columns=["Country Name"], inplace=True)
spain_unemployment_es_melted.drop(columns=["Country Name"], inplace=True)
spain_tourist_number_melted.drop(columns=["Country Name"], inplace=True)
spain_offences_number_melted.drop(columns=["Country Name"], inplace=True)

# Merging the data
merged_data = pd.merge(spain_eurostat_immigrations_melted, spain_unemployment_wb_melted, on="Year", how="outer")
merged_data = pd.merge(merged_data, spain_unemployment_es_melted, on="Year", how="outer")
merged_data = pd.merge(merged_data, spain_tourist_number_melted, on="Year", how="outer")
merged_data = pd.merge(merged_data, spain_offences_number_melted, on="Year", how="outer")

# Using melt to convert the data to a longer format
long_format = merged_data.melt(id_vars=["Year"], var_name="Category", value_name="Value")

# Using pivot to convert the data to have categories as rows and years as columns
pivot_data = long_format.pivot(index="Category", columns="Year", values="Value").reset_index()

# Display the pivot_data to ensure it is shown
pivot_data
```
